## Charger les Donées

```python
# Importer les librairies nécessaires

import pandas as pd
```

```python
# Charger le csv

path = r'C:\Users\shree\Documents\Professional\Job Search 25-26\Interview Prep\PSG\Case Study - PSG\Case Study - PSG\Work\Data\Analysis'
file = r'rfm.csv'

rfm = pd.read_csv(r'{}\{}'.format(path, file), sep = '\t', encoding = 'utf-16')

rfm.describe()
```

```python
rfm.dtypes
```

```python
rfm.head()
```

```python
# Supprimer les ',' en Monetary Value et Recency
# Changer les datatypes au int64

columns = ['Recency', 'Monetary Value']

for col in columns:
    rfm[col] = rfm[col].str.replace(',', '')
    rfm[col] = rfm[col].astype('Int64')

rfm.describe()
```

```python
# Scorer la récence, fréquence, et valeur monétaire des clients en quartiles

labels = [1, 2, 3, 4]

rfm['recency_score'] = pd.qcut(rfm['Recency'], q = 4, labels = labels[::-1])
rfm['frequency_score'] = pd.cut(rfm['Frequency'], bins = [0, 1, 2, 5, 99], labels = [1, 2, 3, 4])
rfm['monetary_score'] = pd.qcut(rfm['Monetary Value'], q = 4, labels = labels)

rfm.tail()
```

```python
# Changer nom de la 1e colonne

rfm.rename({'Unnamed: 0':'fan_id'}, axis = 1, inplace = True)
rfm.head()
```

```python
recency_score = rfm['recency_score']
frequency_score = rfm['frequency_score']
monetary_score = rfm['monetary_score']

rfm_score = []

for i in range(len(recency_score)):
    score = r'{}{}{}'.format(recency_score[i], frequency_score[i], monetary_score[i])
    rfm_score.append(score)

rfm['rfm_score'] = rfm_score

rfm.head()
```

## Exporter les Données

```python
# Selectionner uniquement les colonnes nécessaires : fan_id et rfm_score

rfm_final = rfm[['fan_id', 'rfm_score']]
```

```python
rfm_final.to_csv(r'C:\Users\shree\Documents\Professional\Job Search 25-26\Interview Prep\PSG\Case Study - PSG\Case Study - PSG\Work\Data\Analysis\rfm_scored.csv')
```
